```
import matplotlib.pyplot as plt
import numpy as np
```

```
def f(x):
    return np.cos(x)

res = 100

x = np.linspace(-10, 10, num=res)

y = f(x)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```
def f(x):
    return np.e**x

res = 100

x = np.linspace(-1, 1, num=res)

y = f(x)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```
def f(x):
    return np.log2(x)

res = 100

x = np.linspace(0.01, 4, num=res)

y = f(x)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```
def H(x):
    y = np.zeros(len(x))
    for idx, x in enumerate(x):
        if x >= 0:
            y[idx] = 1
    return y

res = 100

x = np.linspace(-10, 10, num=res)

y = H(x)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```
